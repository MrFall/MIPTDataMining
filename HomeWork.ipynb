```python
#Python №3
def Qsort (arr):
	m_1 = []
	m_2 = []
	m_3 = []
	if len(arr) > 1:
		pivot = arr[0]
		for i in arr:
			if i < pivot:
				m_1.append(i)
			elif i == pivot:
				m_2.append(i)
			elif i > pivot:
				m_3.append(i)
		return Qsort(m_1)+ m_2 + Qsort(m_3)
	else:
		return arr
Qsort([23,15,67,32,50])
```

```python
#NumPy №1
import numpy as np
def fun(a, b, count, d):
	step = ((float(b) - float(a))/(count - 1)) 
	for i in range(0,count):
		d.append(a + i * step)
	return d
	
%timeit fun(1, 100000, 170, [])
```

```python
#Matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
box = [-1, 0, 1.7, 2]
ax = fig.add_axes(box)
x = np.arange(-8.05, 8.0, 0.05)
plt.axis([-2, 2.5, 0, 3])
grapf1, = ax.plot(x, np.sin(x * np.pi), label = 'sin x', color = 'green')
grapf2, = ax.plot(x, np.cos(x * np.pi), label = 'cos x', color = 'blue')
grapf3, = ax.plot(x + 0.25, (x**4 - 3*x**2 + 2), label = 'x^4 - 3*x^2 + 2', color = 'red')
legend = plt.legend(handles = [grapf1, grapf2, grapf3], bbox_to_anchor = (0.6, 1))

x = np.arange(-5.05, 5.0, 0.05)
box_1 = [1, 0, 1.5, 2]
ax_1 = fig.add_axes(box_1)
plt.axis([-8, 8, -5, 5])
grapf5, = ax_1.plot(x, np.abs(x), label = '|x|', color = 'pink') 
grapf6, = ax_1.plot(x, -np.abs(x), label = '-|x|', color = 'purple')
second_legend = plt.legend(handles = [grapf5, grapf6], bbox_to_anchor = (0.6, 1))
plt.savefig('Picture', bbox_inches = 'tight', pad_inches = 1)
plt.show()
```
